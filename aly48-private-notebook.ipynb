```python
#Data Link:
#https://data.wprdc.org/dataset/uniform-crime-reporting-data
import pandas as pd
crime_reports = pd.read_csv("Blotter_Data.csv")
dictionary = {}
for index, row in crime_reports.iterrows():
    if row['INCIDENTNEIGHBORHOOD'] in dictionary:
        dictionary[row['INCIDENTNEIGHBORHOOD']] = dictionary[row['INCIDENTNEIGHBORHOOD']] + 1
    else:
        dictionary[row['INCIDENTNEIGHBORHOOD']] = 1
for i in dictionary:
    print(i, ":", dictionary[i])
```

```python
#Data Link
#https://data.wprdc.org/dataset/station-locations
import fpsnippets as f
import pandas as pd
pogoh = pd.read_csv("Pogoh.csv")
dictionary = {}
for index, row in pogoh.iterrows():
    if f.geo_to_neighborhood(float(row['Latitude']), float(row['Longitude'])) in dictionary:
        dictionary[f.geo_to_neighborhood(float(row['Latitude']), float(row['Longitude']))] = dictionary[f.geo_to_neighborhood(float(row['Latitude']), float(row['Longitude']))] + row['Total Docks']
    else:
        dictionary[f.geo_to_neighborhood(float(row['Latitude']), float(row['Longitude']))] = row['Total Docks']
for i in dictionary:
    print(i, ":", dictionary[i])
```

```python
#Data is taken from arrests csv and Blotter data csv
import pandas as pd
crime_reports = pd.read_csv("Blotter_Data.csv")
arrests = pd.read_csv("arrests.csv")
dictionary_b = {}
dictionary_a = {}
dictionary_c = {}
for index, row in crime_reports.iterrows():
    if row['INCIDENTNEIGHBORHOOD'] in dictionary_b:
        dictionary_b[row['INCIDENTNEIGHBORHOOD']] = dictionary_b[row['INCIDENTNEIGHBORHOOD']] + 1
    else:
        dictionary_b[row['INCIDENTNEIGHBORHOOD']] = 1
for index, row in arrests.iterrows():
    if row['INCIDENTNEIGHBORHOOD'] in dictionary_a:
        dictionary_a[row['INCIDENTNEIGHBORHOOD']] = dictionary_a[row['INCIDENTNEIGHBORHOOD']] + 1
    else:
        dictionary_a[row['INCIDENTNEIGHBORHOOD']] = 1  
    
#calculates the ratios for the metric of arrests/blotter reports    
for i in dictionary_a:
    dictionary_c[i] = float(dictionary_a[i]/dictionary_b[i])

#creating a list that holds the top 25 neighborhoods
dc = dict(sorted(dictionary_c.items(), key=lambda x:x[1], reverse=True))
tp_25 = []
count = 0
for i in dc:
    if count<25:
        tp_25.append(i)
        count = count + 1
    print(i, ":", dc[i])
print(tp_25)
```
